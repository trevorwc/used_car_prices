```python
# Before we do the pre-processing, let's do some feature engineering!
# The categorical variables actually contain information that would be valuable numerically


train['engine_liters'] = train['engine'].str.extract(r'(\d+\.\d+)L').astype(float)
test['engine_liters'] = test['engine'].str.extract(r'(\d+\.\d+)L').astype(float)

train['HP'] = train['engine'].str.extract(r'(\d+\.\d+)HP').astype(float)
test['HP'] = test['engine'].str.extract(r'(\d+\.\d+)HP').astype(float)

train['cylinders'] = train['engine'].str.extract(r'(\d+) Cylinder')
test['cylinders'] = test['engine'].str.extract(r'(\d+) Cylinder')

train['cylinders'] = train['cylinders'].fillna(0).astype(int)
test['cylinders'] = test['cylinders'].fillna(0).astype(int)

train['speed'] = train['transmission'].str.extract(r'(\d+)-Speed')
test['speed'] = test['transmission'].str.extract(r'(\d+)-Speed')

def manual(string):
    if 'Auto' or "A/T" in string:
        return 0
    elif 'Manual' in string:
        return 1
    else:
        return NaN

train['manual'] = train['transmission'].apply(manual)
test['manual'] = test['transmission'].apply(manual)

def accidents(string):
    if isinstance(string, str):
        if 'None' in string:
            return 0 
        else:
            return 1
    

train['accident_bin'] = train['accident'].apply(accidents)
test['accident_bin'] = test['accident'].apply(accidents)

def title(string):
    if isinstance(string, str):
        if "Yes" in string:
            return 1
        else: 
            return 0

train["clean_title_bin"] = train["clean_title"].apply(title)
test["clean_title_bin"] = test["clean_title"].apply(title)
```

```python
X_train, X_val, y_train, y_val = train_test_split(train3[finalvars], train3['price'], test_size=0.2, random_state=42)
```

```python
X_val = X_val.apply(pd.to_numeric, errors='coerce').dropna()
y_val = pd.to_numeric(y_val, errors='coerce').dropna()

X_train = X_train.apply(pd.to_numeric, errors='coerce').dropna()
y_train = pd.to_numeric(y_train, errors='coerce').dropna()
```

```python
y_val
```

```python
# best subset selection
def processSubset(feature_set):
    # Ensure X_val is a DataFrame of numeric values
    X_subset = X_val[list(feature_set)]
    
    # Fit model on feature_set and calculate RSS
    model = sm.OLS(y_val, X_subset).fit()
    RSS = ((model.predict(X_subset) - y_val) ** 2).sum()
    return {"model": model,"features":feature_set, "RSS": RSS}

def getBest(k):
    tic = time.time()
    
    results = []
    
    # Iterate over all combinations of k predictors
    for combo in itertools.combinations(X_val.columns, k):
        results.append(processSubset(combo))
    
    # Wrap everything up in a DataFrame
    models = pd.DataFrame(results)
    
    # Choose the model with the lowest RSS
    best_model = models.loc[models['RSS'].idxmin()]
    
    toc = time.time()
    print(f"Processed {models.shape[0]} models with {k} predictors in {toc - tic:.4f} seconds.")
    
    # Return the best model, along with other information
    return best_model

# Create an empty DataFrame to store best models
models_best = pd.DataFrame(columns=["RSS","features", "model"])

tic = time.time()

# Loop through the number of predictors
for i in range(1, X_val.shape[1] + 1):  # Including the full number of predictors
    best_model = getBest(i)
    models_best.loc[i] = [best_model['RSS'],best_model['features'], best_model['model']]

toc = time.time()
print(f"Total elapsed time: {toc - tic:.4f} seconds.")
```
